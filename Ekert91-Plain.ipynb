```python
from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_gate
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Operator
from qiskit.quantum_info import Statevector
from qiskit import transpile 
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit import ControlledGate
import math 

root2 = math.sqrt(2)

denom1 = math.sqrt(4 + 2*root2)
denom2 = math.sqrt(4 - 2*root2) 


A2B1_transform = [ [ -1 / denom1 , (1 + root2) / denom1 ],
                        [  1 / denom2 , (root2 - 1) / denom2 ] ]
A2B1_transform = Operator(A2B1_transform) 


B3_transform = [ [  1 / denom1 , (1 + root2) / denom1 ],
                        [ -1 / denom2 , (root2 - 1) / denom2 ] ]
B3_transform = Operator(B3_transform) 

def Z(circuit, i):
    pass 

def W(circuit, i):
    circuit.unitary(A2B1_transform,[i])

def X(circuit, i):
    circuit.h(0)

def V(circuit, i):
    circuit.unitary(B3_transform,[i])

basis = [X, W, Z, V]

def random():
    circuit = QuantumCircuit(2)
    circuit.initialize([1/math.sqrt(3), math.sqrt(2)/math.sqrt(3)], 0) 
    circuit.initialize([1/math.sqrt(2), 1/math.sqrt(2)], 1) 
    circuit.measure_all()
    value = get_results(circuit)
    if value[1] == '0':
        return 1
    elif value[0] == '0':
        return 2
    else:
        return 3

def entangled():
    circuit = QuantumCircuit(2,2)
    circuit.h(0)
    circuit.z(0)
    circuit.cx(0, 1)
    circuit.x(0)
    return circuit

def get_results(circuit):
    backend = BasicSimulator()
    qc_compiled = transpile(circuit, backend)
    return next(iter(backend.run(qc_compiled, shots=1).result().get_counts(qc_compiled)))

def get_measurement(key, P):
    ab = random()
    bb = random()
    circuit = entangled()
    basis[ab-1](circuit,0)
    basis[bb](circuit,1)
    circuit.measure([0,1], [0,1])
    get_key_P(ab, bb, key, P, get_results(circuit))

def get_bases(N):
    key = []
    P = {}
    for i in range(9*N//2):
        get_measurement(key, P)
    return key, P

def get_key_P(ab, bb, key, P, message):
    if ab == bb-1:
        key.append(message[0])
    if (ab, bb) not in P:
        P[(ab, bb)] = {("00"):0, ("01"):0, ("10"):0, ("11"):0}
    P[(ab, bb)][message] += 1
    return key, P
        

def get_S(P):
    S = 0
    for i, j in zip ([1,3,1,3], [1,1,3,3]):
        count, E = 0, 0
        if (i, j) in P:
            for k in P[(i, j)]:
                count += P[(i, j)][k]
                if k == "00" or k == "11":
                    if i == 1 and j==3:E -= P[(i, j)][k]
                    else:E += P[(i, j)][k]
                else:
                    if i == 1 and j==3:E += P[(i, j)][k]
                    else:E -= P[(i, j)][k]
        if count != 0: 
            S += (E/count)
    return abs(S)

key, P = get_bases(100)
print(key, get_S(P))    
 
            
```
